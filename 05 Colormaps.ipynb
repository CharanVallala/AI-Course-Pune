```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png',0)
```

```python
plt.imshow(img, cmap = 'gray')
```

Sequential Colormaps

```python
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('gray')
plt.imshow(img, cmap = 'gray')
plt.subplot(2,4,2)
plt.title('Blues')
plt.imshow(img, cmap = 'Blues')
plt.subplot(2,4,3)
plt.title('greens')
plt.imshow(img, cmap = 'Greens')
plt.subplot(2,4,4)
plt.title('Reds')
plt.imshow(img, cmap = 'Reds')
plt.subplot(2,4,5)
plt.title('Oranges')
plt.imshow(img, cmap = 'Oranges')
plt.subplot(2,4,6)
plt.title('PuRd')
plt.imshow(img, cmap = 'PuRd')
plt.subplot(2,4,7)
plt.title('BuGn')
plt.imshow(img, cmap = 'BuGn')
plt.subplot(2,4,8)
plt.title('YlGnBu')
plt.imshow(img, cmap = 'YlGnBu')
```

Diverging colormaps

```python
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('PiYG')
plt.imshow(img, cmap = 'PiYG')
plt.subplot(2,4,2)
plt.title('PRGn')
plt.imshow(img, cmap = 'PRGn')
plt.subplot(2,4,3)
plt.title('BrBG')
plt.imshow(img, cmap = 'BrBG')
plt.subplot(2,4,4)
plt.title('PuOr')
plt.imshow(img, cmap = 'PuOr')
plt.subplot(2,4,5)
plt.title('RdGy')
plt.imshow(img, cmap = 'RdGy')
plt.subplot(2,4,6)
plt.title('RdBu')
plt.imshow(img, cmap = 'RdBu')
plt.subplot(2,4,7)
plt.title('RdYlBu')
plt.imshow(img, cmap = 'RdYlBu')
plt.subplot(2,4,8)
plt.title('Spectral')
plt.imshow(img, cmap = 'Spectral')
```
