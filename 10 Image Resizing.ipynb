```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png')
resized_img = cv2.resize(img,(100,200))
cv2.imshow('resized image',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_img2 = cv2.resize(img,(700,800))
cv2.imshow('resized image2',resized_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

ReScaling without changing aspect ratio

```python
img = cv2.imread('images/lenna.png')
resized_img = cv2.resize(img,None,fx=1.75,fy=1.75)
cv2.imshow('resized image',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from skimage.transform import resize
import matplotlib.pyplot as plt
image = plt.imread("images/lenna.png")
#resize with bilinear interpolation
#resized_img = resize(image, (1000,1200), anti_aliasing = False)
#resize with bilinear interpolation
resized_img = resize(image, (1000,1200), anti_aliasing = True, order=3)
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_img)
```
