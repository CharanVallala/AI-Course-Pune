```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png')
flip_hor = cv2.flip(img,0)
flip_ver = cv2.flip(img,1)
flip_diag = cv2.flip(img,-1)
cv2.imshow('original',img)
cv2.imshow('flip_hor',flip_hor)
cv2.imshow('flip_ver',flip_ver)
cv2.imshow('flip_diag',flip_diag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_hor_rgb = cv2.cvtColor(flip_hor,cv2.COLOR_BGR2RGB)
flip_ver_rgb = cv2.cvtColor(flip_ver,cv2.COLOR_BGR2RGB)
flip_diag_rgb = cv2.cvtColor(flip_diag,cv2.COLOR_BGR2RGB)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=[16,9])
plt.subplot(1,4,1)
plt.imshow(img_rgb)
plt.subplot(1,4,2)
plt.imshow(flip_hor_rgb)
plt.subplot(1,4,3)
plt.imshow(flip_ver_rgb)
plt.subplot(1,4,4)
plt.imshow(flip_diag_rgb)
```
