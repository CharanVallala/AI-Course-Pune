```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1=cv2.imread('images/lenna.png')
img2=cv2.imread('images/sun.png')
```

```python
img1.shape, img2.shape
```

```python
added_image=cv2.add(img1,img2)
cv2.imshow('added_image',added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img3 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
added = cv2.cvtColor(added_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img3)
plt.subplot(1,3,2)
plt.imshow(img4)
plt.subplot(1,3,3)
plt.imshow(added)
```

Opacity

```python
alpha = 0.55 #transparency factor
blended_img = cv2.addWeighted(img3,alpha,img4,1-alpha,0)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img3)
plt.subplot(1,3,2)
plt.imshow(img4)
plt.subplot(1,3,3)
plt.imshow(blended_img)
```

Subtract

```python
sub_image=cv2.subtract(img1,img2)
cv2.imshow('sub_image',sub_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Multiply

```python
mul_image = cv2.multiply(img1,1.5)
img3 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
mul_image_rgb = cv2.cvtColor(mul_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(img3)
plt.subplot(1,2,2)
plt.imshow(mul_image_rgb)
```

Logical

```python
img5=cv2.imread('images/input1.png')
img6=cv2.imread('images/input2.png')
and_img = cv2.bitwise_and(img5,img6,mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img5)
plt.subplot(1,3,2)
plt.imshow(img6)
plt.subplot(1,3,3)
plt.imshow(and_img)
```

```python
img5=cv2.imread('images/input1.png')
img6=cv2.imread('images/input2.png')
or_img = cv2.bitwise_or(img5,img6,mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img5)
plt.subplot(1,3,2)
plt.imshow(img6)
plt.subplot(1,3,3)
plt.imshow(or_img)
```

```python
img5=cv2.imread('images/input1.png')
img6=cv2.imread('images/input2.png')
xor_img = cv2.bitwise_xor(img5,img6,mask = None)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img5)
plt.subplot(1,3,2)
plt.imshow(img6)
plt.subplot(1,3,3)
plt.imshow(xor_img)
```

```python
img5=cv2.imread('images/input1.png')
img6=cv2.imread('images/input2.png')
not_img = cv2.bitwise_not(img6)
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(img6)
plt.subplot(1,2,2)
plt.imshow(not_img)
```
