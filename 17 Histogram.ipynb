```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

num = [np.random.randint(0,100) for x in range(1000)]
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num, bins=range(10,101,10))
```

```python
img1=cv2.imread('images/lenna.png',0) #gray scale 0 reading

plt.figure(figsize=(16,9))
plt.title('Image Histogram')
plt.xlabel('Pixel Values')
plt.ylabel('Count')
plt.xticks(range(0,261,10))
plt.grid()
plt.hist(img1.flatten(), bins=range(0,261,10));
```
