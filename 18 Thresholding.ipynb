```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1=cv2.imread('images/lenna.png',0)
```

```python
new_img = img1 > 127
inv_img = img1 < 127
```

```python
new_img, inv_img
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original')
plt.imshow(img1, cmap='gray')
plt.subplot(132)
plt.title('Binary image')
plt.imshow(new_img, cmap='gray')
plt.subplot(133)
plt.title('Inv Binary image')
plt.imshow(inv_img, cmap='gray')
```

Otsu thresholding

```python
thresh_val, thresh_otsu = cv2.threshold(img1, 0, 255, cv2.THRESH_OTSU)
thresh_val
```

```python
otsu_img = img1 > thresh_val
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original')
plt.imshow(img1, cmap='gray')
plt.subplot(142)
plt.title('Binary image')
plt.imshow(new_img, cmap='gray')
plt.subplot(143)
plt.title('Inv Binary image')
plt.imshow(inv_img, cmap='gray')
plt.subplot(144)
plt.title('Otsu Threshold image')
plt.imshow(otsu_img, cmap='gray')
```

```python
img2=cv2.imread('images/sachin.jpg',0)
new_img = img2 > 127
inv_img = img2 < 127
thresh_val, thresh_otsu = cv2.threshold(img2, 0, 255, cv2.THRESH_OTSU)
otsu_img = img2 > thresh_val
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original')
plt.imshow(img2, cmap='gray')
plt.subplot(142)
plt.title('Binary image')
plt.imshow(new_img, cmap='gray')
plt.subplot(143)
plt.title('Inv Binary image')
plt.imshow(inv_img, cmap='gray')
plt.subplot(144)
plt.title('Otsu Threshold image')
plt.imshow(otsu_img, cmap='gray')
```
