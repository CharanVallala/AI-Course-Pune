```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1=cv2.imread('images/lenna.png')
```

Sobel Edge Detection

```python
#Sobel paarameters : ksize, dx, dy, depth
sobelx = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize=3)
#sobel for horizontal edges
sobely = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize=3)
#sobel for vertical edges

#combine the sobel images-optional
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img = np.hstack([sobelx,sobely, combined_sobel])
cv2.imshow('all images',final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img2=cv2.imread('images/lenna.png',0)
#Sobel paarameters : ksize, dx, dy, depth
sobelx = cv2.Sobel(img2, cv2.CV_64F, 1, 0, ksize=3)
#sobel for horizontal edges
sobely = cv2.Sobel(img2, cv2.CV_64F, 0, 1, ksize=3)
#sobel for vertical edges

#combine the sobel images-optional
combined_sobel = cv2.add(sobelx, sobely)
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original')
plt.imshow(img2, cmap='gray')
plt.subplot(142)
plt.title('SobelX')
plt.imshow(abs(sobelx), cmap='gray')
plt.subplot(143)
plt.title('SobelY')
plt.imshow(abs(sobely), cmap='gray')
plt.subplot(144)
plt.title('Combined Sobel')
plt.imshow(abs(combined_sobel), cmap='gray')
```

Canny Edge Detection

```python
img2=cv2.imread('images/circle.png',0)
edges = cv2.Canny(img2,50,100) #adjust threshold if needed
cv2.imshow('original',img2)
cv2.imshow('canny edge image',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Canny is most effective algorithm to detect edges
