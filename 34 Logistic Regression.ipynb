Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
pd.DataFrame(x_scaled).describe()
```

Cross Validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

Build the model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
#train the algorithm
classifier.fit(x_train, y_train)
```

Predict on the unseen  data

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
#If Feature Scaling is not used, Salary being a large number than age, dominates the result, showed all 0's.
```

Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
#67 zeros, 22 ones are predicted correctly, hence 89% accuracy
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm.diagonal().sum() / cm.sum()
```

```python
print(classification_report(y_test, y_pred))
```

```python
#recall is individual class accuracy : class 0 accuracy is 99%, class 1 accuracy is 69%, which is not acceptable 
#hence this model is inefficient though overall accuracy  is 89%
```

```python
#accuracy of class 0 / sensitivity of algorithm
r0 = 67/68
r0
```

```python
#accuracy of class 1 / specificity of algorithm
r1 = 22/32
r1
```

```python
#precision-0 /Positive predictive value
p0 = 67/77
p0
```

```python
#precision-1 /Negative predictive value
p1 = 22/23
p1
```

```python
# f1-score - 0
2*p0*r0/(p0+r0)
```

```python
# f1-score - 1
2*p1*r1/(p1+r1)
```
