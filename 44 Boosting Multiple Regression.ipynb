```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('../datasets/mtcars.csv')
df
```

Seperate the input and output variables

```python
#disp, hp, wt
x = df[['disp','hp','wt']]
x
```

```python
#mpg
y = df['mpg']
y
```

```python
x.corrwith(y) #output is always one dimensional, input is multi dimesnional
```

build the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
```

```python
regressor.score(x,y)
```

```python
regressor.coef_ #hence negative regression
```

```python
regressor.intercept_
```

Predict on unseen data

```python
disp = 221
hp = 102
wt = 2.91
```

```python
help(regressor.predict)
```

```python
regressor.predict([[disp,hp,wt]])
```

```python
new = pd.DataFrame({
    'disp': [221],
    'hp': [102],
    'wt': [2.91]
})
new
```

```python
regressor.predict(new)
```

```python
(disp * regressor.coef_[0]) + (hp * regressor.coef_[1]) + (wt * regressor.coef_[1]) + (regressor.intercept_)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.xlabel('disp')
plt.ylabel('mpg')
plt.scatter(df['disp'],df['mpg'])
plt.subplot(1,3,2)
plt.xlabel('hp')
plt.ylabel('mpg')
plt.scatter(df['hp'],df['mpg'])
plt.subplot(1,3,3)
plt.xlabel('wt')
plt.ylabel('mpg')
plt.scatter(df['wt'],df['mpg'])
```

Adaboost the Regression

```python
from sklearn.ensemble import AdaBoostRegressor
reg_adb = AdaBoostRegressor(regressor, n_estimators=3, random_state=0)
reg_adb.fit(x, y)
```

```python
reg_adb.score(x,y)
```
